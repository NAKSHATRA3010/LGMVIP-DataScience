```python
#IMPORT THE REQUIRED LIBRARIES
!pip install opencv-python
import cv2
from google.colab.patches import cv2_imshow
```

```python
# READ THE IMAGE IN RGB FORMAT
image = cv2.imread("sunflower.jpg")
image_resize = cv2.resize(image, (400,400), interpolation = cv2.INTER_LINEAR_EXACT)
cv2_imshow(image_resize)
cv2.waitKey(0)
```

```python
# CONVERT THE ORIGINAL IMAGE INTO GRAYSCALE FORMAT

gray_image = cv2.cvtColor(image_resize,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

```python
#CONVERT THE GRAY_IMAGE INTO INVERTED GRAY_IMAGE

inverted_image = 255- gray_image
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

```python
#CONVERT THE IMAGE INTO BLUR IMAGE

blur_image = cv2.GaussianBlur(inverted_image, (21,21),0)
cv2_imshow(blur_image)
cv2.waitKey(0)
```

```python
#CONVERT THE IMAGE INTO SKETCH FORMAT

sketch = 255 - blur_image

pencil_sketch = cv2.divide(gray_image,sketch,scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
